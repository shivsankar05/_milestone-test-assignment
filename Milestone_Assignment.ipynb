```python
#Q1)What is the difference between static and dynamic variables in python?
# Static vs Dynamic Variables in Python

# Static Variables:
# - Shared by all instances of a class
# - Defined within the class but outside any instance methods

class MyClass:
    static_variable = 42  # Static variable

# Accessing static variable
print(MyClass.static_variable)  # Output: 42

# Dynamic Variables:
# - Unique to each instance
# - Defined within methods (usually `__init__`)

class MyClass:
    def __init__(self, value):
        self.instance_variable = value  # Dynamic variable

# Creating instances
obj1 = MyClass(1)
obj2 = MyClass(2)

# Accessing dynamic variables
print(obj1.instance_variable)  # Output: 1
print(obj2.instance_variable)  # Output: 2
```

```python
#Q2)Explain the purpose of pop , popitem, clear in a dictionary with suitable example?
# Dictionary Methods: pop, popitem, clear

# pop: Removes and returns the value for the specified key
my_dict = {'a': 1, 'b': 2, 'c': 3}
value = my_dict.pop('b')  # Removes key 'b' and returns its value
print(value)  # Output: 2
print(my_dict)  # Output: {'a': 1, 'c': 3}

# popitem: Removes and returns the last inserted key-value pair as a tuple
my_dict = {'a': 1, 'b': 2, 'c': 3}
item = my_dict.popitem()  # Removes and returns the last key-value pair
print(item)  # Output: ('c', 3)
print(my_dict)  # Output: {'a': 1, 'b': 2}

# clear: Removes all items from the dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}
my_dict.clear()  # Clears all items from the dictionary
print(my_dict)  # Output: {}
```

```python
#Q8)What is break, continue and pass in python?
# break: Exits the current loop (for or while) immediately.

print("Using break:")
for i in range(5):
    if i == 3:
        break
    print(i)
# Output: 0, 1, 2

# continue: Skips the rest of the code inside the current loop iteration and moves to the next iteration.

print("Using continue:")
for i in range(5):
    if i == 3:
        continue
    print(i)
# Output: 0, 1, 2, 4

# pass: A placeholder that does nothing. It is used when a statement is syntactically required but you have no code to execute.

print("Using pass:")
for i in range(3):
    if i == 1:
        pass  # Do nothing
    else:
        print(i)
# Output: 0, 2
```

```python
#Q10)What are global, protected and private attributes in python?
"""
- Global Attributes: Variables defined outside of functions or classes and accessible from anywhere in the module.

- Protected Attributes: Variables prefixed with a single underscore (_) to indicate they should not be accessed directly outside of the class.

- Private Attributes: Variables prefixed with a double underscore (__) to make them name-mangled and less accessible from outside the class.
"""
# Example:

# Global attribute
global_var = "I am global"

class Example:
    def __init__(self):
        self._protected_attr = "I am protected"  # Protected attribute
        self.__private_attr = "I am private"     # Private attribute

    def show_attributes(self):
        return (self._protected_attr, self.__private_attr)

# Creating an instance of Example
example = Example()

# Accessing attributes
print(global_var)                  # Output: I am global
print(example.show_attributes())   # Output: ('I am protected', 'I am private')

# Attempting to access protected and private attributes directly
print(example._protected_attr)     # Output: I am protected
try:
    print(example.__private_attr)  # This will raise an AttributeError
except AttributeError as e:
    print(e)  # Output: 'Example' object has no attribute '__private_attr'

# Accessing private attribute using name mangling
print(example._Example__private_attr)  # Output: I am private
```

```python
#Q15)What are decorators in python?Explain with an example, write down its use cases.
"""
**Decorators**: A decorator is a function that modifies or extends the behavior of another function or method. It is applied to functions or methods using the @ symbol above their definition.

**Purpose**:
- To add functionality to functions or methods in a reusable way.
- To modify or enhance the behavior of functions without changing their actual code.

**Example**:
"""
def decorator_function(original_function):
    def wrapper_function():
        print("Decorator added this behavior.")
        original_function()
        print("Decorator added this behavior too.")
    return wrapper_function

@decorator_function
def say_hello():
    print("Hello!")

# Using the decorated function
say_hello()
# Output:
# Decorator added this behavior.
# Hello!
# Decorator added this behavior too.

"""
**Use Cases**:
1. **Logging**: Automatically log information before or after a function execution.
2. **Authorization**: Check if a user has permission before executing a function.
3. **Caching**: Cache results of expensive function calls to improve performance.
4. **Monitoring**: Track performance metrics of functions.
5. **Validation**: Validate inputs or outputs of functions.

"""

# Example Use Case: Logging

def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Function '{func.__name__}' called with arguments: {args}, {kwargs}")
        result = func(*args, **kwargs)
        print(f"Function '{func.__name__}' returned: {result}")
        return result
    return wrapper

@log_decorator
def add(a, b):
    return a + b

# Using the decorated function
add(5, 3)
# Output:
# Function 'add' called with arguments: (5, 3), {}
# Function 'add' returned: 8
```

```python
#Q16)How is memory managed in python?
"""
Memory management in Python involves several mechanisms:

1. **Automatic Memory Management**: Python uses a garbage collector to automatically manage memory, reclaiming unused memory.

2. **Reference Counting**: Python keeps track of the number of references to each object. When the reference count drops to zero, the memory occupied by the object is deallocated.

3. **Garbage Collection**: Python's garbage collector periodically identifies and cleans up cyclic references (objects that reference each other but are no longer reachable from the program).

4. **Memory Pools**: Python uses a memory pool system to manage small objects efficiently. It reduces overhead and fragmentation by allocating memory in blocks.

5. **Dynamic Typing**: Python dynamically allocates and deallocates memory for variables based on their types and usage.

# Example to Demonstrate Reference Counting

import sys

# Create some objects
a = [1, 2, 3]
b = a

# Check reference count
print(sys.getrefcount(a))  # Output: Reference count of list 'a' (usually 2)

# Remove reference
del b
print(sys.getrefcount(a))  # Output: Reference count of list 'a' (usually 1)

# Example to Demonstrate Garbage Collection

import gc

# Force garbage collection
gc.collect()

print("Garbage collection completed.")
"""

# Example: Reference Counting and Garbage Collection

class MyClass:
    def __init__(self, value):
        self.value = value

obj1 = MyClass(10)
obj2 = obj1

# Reference count of obj1
import sys
print(sys.getrefcount(obj1))  # Output: Reference count of obj1

# Remove references
del obj1
del obj2

# Force garbage collection
import gc
gc.collect()
print("Garbage collection completed.")
```

```python
#Q16)What is lambda in python?why is it used?
"""
**Lambda**: A lambda function is a small, anonymous function defined using the `lambda` keyword. It can have any number of arguments but only one expression. The expression is evaluated and returned.

**Purpose**:
- To create small, throwaway functions without the need to define a full function using `def`.
- Often used for short-term use, especially as arguments to higher-order functions like `map()`, `filter()`, and `sorted()`.

**Syntax**:
lambda arguments: expression

# Example of a Lambda Function
"""
# Using a lambda function to add two numbers
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

# Using a lambda function with `map()`
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]

# Using a lambda function with `filter()`
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]

# Using a lambda function with `sorted()` for custom sorting
points = [(1, 2), (3, 1), (5, 0)]
sorted_points = sorted(points, key=lambda p: p[1])
print(sorted_points)  # Output: [(5, 0), (3, 1), (1, 2)]
```

```python
#Q17)Explain split() and join() functions in python.
"""
**split()**: The `split()` method is used to split a string into a list of substrings based on a specified delimiter. By default, it splits by any whitespace.

**Syntax**:
str.split([separator[, maxsplit]])

- `separator`: Optional. Specifies the delimiter to split on. Default is any whitespace.
- `maxsplit`: Optional. Specifies the maximum number of splits. Default is -1, meaning "all occurrences".

**join()**: The `join()` method is used to join elements of an iterable (like a list or tuple) into a single string, with a specified separator.

**Syntax**:
separator.join(iterable)

- `separator`: The string used to join the elements.
- `iterable`: An iterable whose elements are joined into a single string.

# Example of split() and join()

"""
# Example of split()
text = "Hello world, this is Python."
words = text.split()  # Splits on whitespace by default
print(words)  # Output: ['Hello', 'world,', 'this', 'is', 'Python.']

# Example of split() with a separator
csv_text = "apple,banana,cherry"
fruits = csv_text.split(",")  # Splits on comma
print(fruits)  # Output: ['apple', 'banana', 'cherry']

# Example of join()
separator = " - "
joined_text = separator.join(fruits)  # Joins elements with ' - ' as separator
print(joined_text)  # Output: 'apple - banana - cherry'

# Example of join() with a list of words
sentence = " ".join(words)  # Joins elements with space as separator
print(sentence)  # Output: 'Hello world, this is Python.'
```

```python
#Q18)What are iterators, iterable and generators in python?
"""
**Iterable**: An object that implements the `__iter__()` method or has a `__getitem__()` method that can return items one at a time. Iterables can be passed to the `iter()` function to create an iterator.

**Iterator**: An object that implements the `__iter__()` and `__next__()` methods. Iterators produce values one at a time from an iterable and keep track of their state.

**Generator**: A special type of iterator created using a function with the `yield` keyword. Generators are a concise way to write iterators and are lazy, meaning they generate values on-the-fly.

# Examples
"""
# Example of Iterable
class MyIterable:
    def __init__(self):
        self.values = [1, 2, 3]

    def __iter__(self):
        self.index = 0
        return self

    def __next__(self):
        if self.index < len(self.values):
            result = self.values[self.index]
            self.index += 1
            return result
        else:
            raise StopIteration

iterable = MyIterable()
for value in iterable:
    print(value)  # Output: 1 2 3

# Example of Iterator
iterator = iter([1, 2, 3])
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2

# Example of Generator
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
for value in gen:
    print(value)  # Output: 1 2 3
```

```python
#Q20)Pillars of OOPs

# Pillars of Object-Oriented Programming (OOP) in Python
"""
**1. Encapsulation**:
Encapsulation is the concept of wrapping data (attributes) and methods (functions) into a single unit, a class. It hides the internal state of the object from the outside world and only exposes a controlled interface.

**Example**:
"""
class EncapsulatedClass:
    def __init__(self, value):
        self.__value = value  # Private attribute

    def get_value(self):
        return self.__value

    def set_value(self, new_value):
        self.__value = new_value

obj = EncapsulatedClass(10)
print(obj.get_value())  # Output: 10
obj.set_value(20)
print(obj.get_value())  # Output: 20

"""
**2. Abstraction**:
Abstraction involves hiding the complex implementation details and showing only the essential features of an object. It is achieved using abstract classes and methods.

**Example**:
"""

from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def abstract_method(self):
        pass

class ConcreteClass(AbstractClass):
    def abstract_method(self):
        return "Implementation of abstract method"

obj = ConcreteClass()
print(obj.abstract_method())  # Output: Implementation of abstract method

"""
**3. Inheritance**:
Inheritance is the mechanism by which one class (child class) inherits attributes and methods from another class (parent class). It promotes code reusability and establishes a relationship between classes.

**Example**:
"""

class ParentClass:
    def parent_method(self):
        return "This is a parent method"

class ChildClass(ParentClass):
    def child_method(self):
        return "This is a child method"

obj = ChildClass()
print(obj.parent_method())  # Output: This is a parent method
print(obj.child_method())   # Output: This is a child method

"""
**4. Polymorphism**:
Polymorphism allows different classes to be treated as instances of the same class through a common interface. It is typically achieved through method overriding and operator overloading.

**Example**:
"""

class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

def make_animal_speak(animal):
    print(animal.speak())

dog = Dog()
cat = Cat()
make_animal_speak(dog)  # Output: Dog barks
make_animal_speak(cat)  # Output: Cat meows
```

```python
#Q21)How will you check if class is a child class of another class?
"""
In Python, we can check if a class is a child class of another class using the `issubclass()` function.

**Syntax**:
issubclass(class, classinfo)

- `class`: The class to check.
- `classinfo`: The class or a tuple of classes to check against.

**Example**:
"""

class Parent:
    pass

class Child(Parent):
    pass

class Unrelated:
    pass

# Check if Child is a subclass of Parent
print(issubclass(Child, Parent))  # Output: True

# Check if Unrelated is a subclass of Parent
print(issubclass(Unrelated, Parent))  # Output: False

# Check if Child is a subclass of multiple classes
print(issubclass(Child, (Parent, Unrelated)))  # Output: True
```

```python
#Q22)How does inheritance work in python?Explain all types of inheritance with an example.
"""
Inheritance allows one class (child class) to inherit attributes and methods from another class (parent class). It promotes code reuse and establishes a relationship between classes.

**Types of Inheritance**:

1. **Single Inheritance**: A class inherits from a single parent class.
2. **Multiple Inheritance**: A class inherits from more than one parent class.
3. **Multilevel Inheritance**: A class inherits from a parent class, which in turn inherits from another class.
4. **Hierarchical Inheritance**: Multiple classes inherit from a single parent class.
5. **Hybrid Inheritance**: A combination of two or more types of inheritance.

# Examples
"""
# Single Inheritance
class Parent:
    def parent_method(self):
        return "This is a parent method"

class Child(Parent):
    def child_method(self):
        return "This is a child method"

child = Child()
print(child.parent_method())  # Output: This is a parent method
print(child.child_method())   # Output: This is a child method

# Multiple Inheritance
class Base1:
    def method1(self):
        return "Method from Base1"

class Base2:
    def method2(self):
        return "Method from Base2"

class Derived(Base1, Base2):
    pass

derived = Derived()
print(derived.method1())  # Output: Method from Base1
print(derived.method2())  # Output: Method from Base2

# Multilevel Inheritance
class Grandparent:
    def grandparent_method(self):
        return "This is a grandparent method"

class Parent(Grandparent):
    def parent_method(self):
        return "This is a parent method"

class Child(Parent):
    def child_method(self):
        return "This is a child method"

child = Child()
print(child.grandparent_method())  # Output: This is a grandparent method
print(child.parent_method())       # Output: This is a parent method
print(child.child_method())        # Output: This is a child method

# Hierarchical Inheritance
class Parent:
    def parent_method(self):
        return "This is a parent method"

class Child1(Parent):
    def child1_method(self):
        return "This is child1 method"

class Child2(Parent):
    def child2_method(self):
        return "This is child2 method"

child1 = Child1()
child2 = Child2()
print(child1.parent_method())  # Output: This is a parent method
print(child2.parent_method())  # Output: This is a parent method

# Hybrid Inheritance (Example of combining multiple and multilevel inheritance)
class Base:
    def base_method(self):
        return "Method from Base"

class A(Base):
    def a_method(self):
        return "Method from A"

class B(Base):
    def b_method(self):
        return "Method from B"

class C(A, B):
    pass

c = C()
print(c.base_method())  # Output: Method from Base
print(c.a_method())     # Output: Method from A
print(c.b_method())     # Output: Method from B
```

```python
#Q23)What is encapsulation? Explain it with an example.
"""
**Encapsulation**: Encapsulation is the concept of bundling data (attributes) and methods (functions) that operate on the data into a single unit, a class. It also involves restricting direct access to some of an object's components to protect the integrity of the data.

**Purpose**:
- To hide the internal state of an object and only expose a controlled interface.
- To prevent unintended interference and misuse of data.
- To achieve data hiding and abstraction.

**Implementation**:
- **Private Attributes**: Attributes that are meant to be private are prefixed with double underscores (`__`).
- **Public Methods**: Methods that provide a controlled way to access or modify private attributes.

# Example
"""
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds or invalid amount")

    def get_balance(self):
        return self.__balance

# Create an instance of BankAccount
account = BankAccount("Alice", 1000)

# Access and modify balance through methods
print(account.get_balance())  # Output: 1000
account.deposit(500)
print(account.get_balance())  # Output: 1500
account.withdraw(200)
print(account.get_balance())  # Output: 1300

# Attempt to access private attribute directly
print(account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'
```

```python
# Program to accept marks in five subjects, display them, calculate the sum and percentage, and find the grade

# Function to accept and display marks
def accept_and_display_marks():
    subjects = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
    marks = []

    for subject in subjects:
        while True:
            try:
                mark = float(input(f"Enter marks for {subject}: "))
                if 0 <= mark <= 100:
                    marks.append(mark)
                    break
                else:
                    print("Please enter a valid mark between 0 and 100.")
            except ValueError:
                print("Invalid input! Please enter numeric values.")

    # Display the marks
    print("\nMarks obtained:")
    for subject, mark in zip(subjects, marks):
        print(f"{subject}: {mark}")

    return marks

# Function to calculate sum and percentage
def calculate_sum_and_percentage(marks):
    total_marks = sum(marks)
    percentage = total_marks / len(marks)
    return total_marks, percentage

# Function to determine grade based on percentage using match-case
def find_grade(percentage):
    match percentage:
        case _ if percentage > 85:
            return "A"
        case _ if 75 <= percentage <= 85:
            return "B"
        case _ if 50 <= percentage < 75:
            return "C"
        case _ if 30 < percentage < 50:
            return "D"
        case _ if percentage <= 30:
            return "Reappear"

# Main program
marks = accept_and_display_marks()
total_marks, percentage = calculate_sum_and_percentage(marks)

# Display the total marks, percentage, and grade
print(f"\nTotal Marks: {total_marks}")
print(f"Percentage: {percentage:.2f}%")
grade = find_grade(percentage)
print(f"Grade: {grade}")
```

```python
# Base class for Bank Account
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        else:
            print("Invalid withdraw amount or insufficient funds.")

    def balance_inquiry(self):
        print(f"Balance: ${self.balance:.2f}")

# Subclass for Savings Account
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0, interest_rate=0.02):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest applied: ${interest:.2f}")

# Subclass for Checking Account
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0, overdraft_limit=500):
        super().__init__(account_number, account_holder, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        else:
            print("Invalid withdraw amount or exceeds overdraft limit.")

# Test the banking system
def test_banking_system():
    # Create instances of different account types
    savings = SavingsAccount("SA123", "Alice", 1000)
    checking = CheckingAccount("CA123", "Bob", 500)

    # Perform transactions on Savings Account
    print("Savings Account Transactions:")
    savings.deposit(200)
    savings.withdraw(150)
    savings.balance_inquiry()
    savings.apply_interest()
    savings.balance_inquiry()

    print("\nChecking Account Transactions:")
    # Perform transactions on Checking Account
    checking.deposit(300)
    checking.withdraw(700)
    checking.withdraw(200)  # This should fail due to overdraft limit
    checking.balance_inquiry()

# Run the test
test_banking_system()
```

**Q22)Explain PROBABILITY MASS FUNCTION (PMF) and PROBABILITY DENSITY FUNCTION (PDF). and what is the difference between them?**

```python
# Explanation
"""
Probability Mass Function (PMF):
- PMF is used for discrete random variables.
- It gives the probability that a discrete random variable is exactly equal to some value.
- Mathematically, for a discrete random variable X, PMF is P(X=x).

Probability Density Function (PDF):
- PDF is used for continuous random variables.
- It describes the likelihood of a continuous random variable to take on a particular value.
- The probability of the variable falling within a particular range of values is given by the integral of this variable's PDF over that range.
- Mathematically, for a continuous random variable X, the probability that X lies within the interval [a, b] is the integral of the PDF from a to b.

Difference:
- PMF deals with discrete variables while PDF deals with continuous variables.
- PMF gives the probability of a specific outcome, whereas PDF gives the probability density, and probabilities are found by integrating over a range.
"""

# Example Code

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# PMF example with a discrete distribution (Binomial)
n, p = 10, 0.5  # parameters for the binomial distribution
x = np.arange(0, n+1)
pmf = [stats.binom.pmf(k, n, p) for k in x]

# PDF example with a continuous distribution (Normal)
mean, std = 0, 1  # parameters for the normal distribution
x_continuous = np.linspace(-3, 3, 1000)
pdf = stats.norm.pdf(x_continuous, mean, std)

# Plotting PMF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(x, pmf, use_line_collection=True)
plt.title('Probability Mass Function (PMF) - Binomial Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')

# Plotting PDF
plt.subplot(1, 2, 2)
sns.lineplot(x=x_continuous, y=pdf)
plt.title('Probability Density Function (PDF) - Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Explanation
"""
Correlation:
- Correlation measures the strength and direction of a linear relationship between two variables.
- It quantifies the degree to which two variables are related.

Types of Correlation:
1. Positive Correlation:
   - As one variable increases, the other variable also increases.
   - Example: Height and Weight.

2. Negative Correlation:
   - As one variable increases, the other variable decreases.
   - Example: Price and Demand.

3. Zero Correlation:
   - There is no linear relationship between the variables.
   - Example: Shoe size and IQ.

Methods of Determining Correlation:
1. Pearson Correlation Coefficient:
   - Measures the linear relationship between two continuous variables.
   - Values range from -1 to 1.
     - 1: Perfect positive correlation.
     - -1: Perfect negative correlation.
     - 0: No correlation.
   - Formula: r = Σ[(X - X̄)(Y - Ȳ)] / sqrt[Σ(X - X̄)² Σ(Y - Ȳ)²]

2. Spearman's Rank Correlation:
   - Measures the strength and direction of the relationship between two ranked variables.
   - Values range from -1 to 1.
   - Non-parametric measure, used for ordinal data.
   - Formula: ρ = 1 - (6 Σ d_i^2) / [n(n^2 - 1)]
     - d_i: Difference between the ranks of each observation.
     - n: Number of observations.

3. Kendall's Tau Correlation:
   - Measures the ordinal association between two variables.
   - Values range from -1 to 1.
   - Used for small sample sizes and non-parametric data.
   - Formula: τ = (number of concordant pairs - number of discordant pairs) / [n(n-1)/2]

Visualization:
- Scatter plots are often used to visualize the relationship between two variables.
"""

# Example Code

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, kendalltau

# Generate sample data
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y_positive = 2*x + np.random.normal(0, 0.1, 100)  # Positive correlation
y_negative = -2*x + np.random.normal(0, 0.1, 100) # Negative correlation
y_zero = np.random.normal(0, 1, 100)              # Zero correlation

# Compute correlation coefficients
pearson_corr_pos = pearsonr(x, y_positive)[0]
pearson_corr_neg = pearsonr(x, y_negative)[0]
pearson_corr_zero = pearsonr(x, y_zero)[0]

spearman_corr_pos = spearmanr(x, y_positive)[0]
spearman_corr_neg = spearmanr(x, y_negative)[0]
spearman_corr_zero = spearmanr(x, y_zero)[0]

kendall_corr_pos = kendalltau(x, y_positive)[0]
kendall_corr_neg = kendalltau(x, y_negative)[0]
kendall_corr_zero = kendalltau(x, y_zero)[0]

# Plotting
plt.figure(figsize=(18, 5))

# Positive Correlation
plt.subplot(1, 3, 1)
sns.scatterplot(x=x, y=y_positive)
plt.title(f'Positive Correlation\nPearson: {pearson_corr_pos:.2f}, Spearman: {spearman_corr_pos:.2f}, Kendall: {kendall_corr_pos:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

# Negative Correlation
plt.subplot(1, 3, 2)
sns.scatterplot(x=x, y=y_negative)
plt.title(f'Negative Correlation\nPearson: {pearson_corr_neg:.2f}, Spearman: {spearman_corr_neg:.2f}, Kendall: {kendall_corr_neg:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

# Zero Correlation
plt.subplot(1, 3, 3)
sns.scatterplot(x=x, y=y_zero)
plt.title(f'Zero Correlation\nPearson: {pearson_corr_zero:.2f}, Spearman: {spearman_corr_zero:.2f}, Kendall: {kendall_corr_zero:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

**Q31)The mean of a distribution is 60 with a standard deviation of 10. Assuming that the distribution is normal, what percentage of items be**

(i) between 60 and 72,

(ii) between 50 and 60,

(iii) beyond 72 and

(iv) between
70 and 80?

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Given values
mean = 60
std_dev = 10

# Calculate Z-scores and percentages
def calculate_percentage(mean, std_dev, lower, upper=None):
    if upper is None:
        z_lower = (lower - mean) / std_dev
        percentage = 1 - stats.norm.cdf(z_lower)
    else:
        z_lower = (lower - mean) / std_dev
        z_upper = (upper - mean) / std_dev
        percentage = stats.norm.cdf(z_upper) - stats.norm.cdf(z_lower)
    return percentage * 100

# Calculations
percent_60_72 = calculate_percentage(mean, std_dev, 60, 72)
percent_50_60 = calculate_percentage(mean, std_dev, 50, 60)
percent_beyond_72 = calculate_percentage(mean, std_dev, 72)
percent_70_80 = calculate_percentage(mean, std_dev, 70, 80)

# Results
results = {
    "60 and 72": percent_60_72,
    "50 and 60": percent_50_60,
    "beyond 72": percent_beyond_72,
    "70 and 80": percent_70_80
}

# Display results
for key, value in results.items():
    print(f"Percentage of items between {key}: {value:.2f}%")
```

**Q35)A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the hypothesis that the population standard derivation is 10.5. Hint(use chi- square distribution).**

```python
import scipy.stats as stats

# Given data
n = 25  # Sample size
s = 9.0  # Sample standard deviation
sigma = 10.5  # Hypothesized population standard deviation
alpha = 0.05  # Significance level

# Calculate the chi-square test statistic
chi_square_statistic = ((n - 1) * s**2) / sigma**2

# Degrees of freedom
df = n - 1

# Critical values for the chi-square distribution
critical_value_low = stats.chi2.ppf(alpha / 2, df)
critical_value_high = stats.chi2.ppf(1 - alpha / 2, df)

# Print results
print(f"Chi-Square Statistic: {chi_square_statistic:.2f}")
print(f"Critical Value Low: {critical_value_low:.2f}")
print(f"Critical Value High: {critical_value_high:.2f}")

# Decision
if chi_square_statistic < critical_value_low or chi_square_statistic > critical_value_high:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Q) Explain the bias-variance tradeoff.

# ## Explanation
#
# - **Bias**: Error due to overly simplistic assumptions in the learning algorithm. High bias can cause the model to miss relevant relations between features and target outputs (underfitting).
# - **Variance**: Error due to too much complexity in the learning algorithm. High variance can cause the model to model the random noise in the training data (overfitting).
# - **Tradeoff**: The goal is to find a balance between bias and variance to minimize the total error.

# Visualizing Bias-Variance Tradeoff
import matplotlib.pyplot as plt
import numpy as np

# Error values
bias = np.linspace(0, 1, 100)
variance = 1 - bias
total_error = bias + variance

plt.figure(figsize=(10, 6))
plt.plot(bias, label="Bias")
plt.plot(variance, label="Variance")
plt.plot(total_error, label="Total Error")
plt.xlabel("Model Complexity")
plt.ylabel("Error")
plt.legend()
plt.title("Bias-Variance Tradeoff")
plt.show()

# Q) What are precision and recall? How are they different from accuracy?

# ## Explanation
#
# - **Precision**: The ratio of correctly predicted positive observations to the total predicted positives. Precision = TP / (TP + FP)
# - **Recall**: The ratio of correctly predicted positive observations to all observations in the actual class. Recall = TP / (TP + FN)
# - **Accuracy**: The ratio of correctly predicted observations to the total observations. Accuracy = (TP + TN) / (TP + TN + FP + FN)
# - **Difference**: Accuracy measures the overall correctness, while precision and recall focus on the performance on positive class predictions.

# Q) What is overfitting and how can it be prevented?

# ## Explanation
#
# - **Overfitting**: When a model performs well on training data but poorly on new, unseen data. It captures noise in the training data.
# - **Prevention**:
#   - Use more training data
#   - Simplify the model
#   - Use regularization techniques
#   - Cross-validation
#   - Pruning (in decision trees)

# Q) Explain the concept of cross-validation.

# ## Explanation
#
# - **Cross-Validation**: A technique to assess the performance of a model. It involves splitting the dataset into multiple folds, training on some folds, and validating on the remaining fold(s). This process is repeated several times.
# - **Common Types**:
#   - k-Fold Cross-Validation
#   - Leave-One-Out Cross-Validation
#   - Stratified k-Fold Cross-Validation

# Example of k-Fold Cross-Validation
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target
model = RandomForestClassifier()
scores = cross_val_score(model, X, y, cv=5)
print("Cross-Validation Scores:", scores)
print("Mean Cross-Validation Score:", np.mean(scores))

# Q) What is the difference between a classification and a regression problem?

# ## Explanation
#
# - **Classification**: Predicts a discrete label or category. Example: Spam detection (spam/not spam).
# - **Regression**: Predicts a continuous value. Example: Predicting house prices.

# Q) Explain the concept of ensemble learning.

# ## Explanation
#
# - **Ensemble Learning**: Combines multiple machine learning models to improve performance. The idea is that a group of weak learners can form a strong learner.
# - **Common Methods**:
#   - Bagging (e.g., Random Forest)
#   - Boosting (e.g., AdaBoost, Gradient Boosting)
#   - Stacking

# Q) What is gradient descent and how does it work?

# ## Explanation
#
# - **Gradient Descent**: An optimization algorithm used to minimize the cost function. It iteratively adjusts the model parameters to find the minimum cost.
# - **Process**:
#   1. Initialize parameters.
#   2. Calculate the gradient of the cost function with respect to the parameters.
#   3. Update parameters by moving in the direction of the negative gradient.
#   4. Repeat steps 2-3 until convergence.

# Q) Describe the difference between batch gradient descent and stochastic gradient descent.

# ## Explanation
#
# - **Batch Gradient Descent**: Calculates the gradient using the entire dataset. More accurate but slower for large datasets.
# - **Stochastic Gradient Descent (SGD)**: Calculates the gradient using a single data point. Faster but noisier updates.

# Q) What is the Curse of Dimensionality in machine learning?

# ## Explanation
#
# - **Curse of Dimensionality**: Refers to various phenomena that arise when analyzing and organizing data in high-dimensional spaces. It becomes difficult to visualize, compute, and make accurate predictions as dimensions increase.

# Q) Explain the difference between L1 and L2 regularization.

# ## Explanation
#
# - **L1 Regularization (Lasso)**: Adds the absolute value of the magnitude of the coefficient as a penalty term to the loss function. Encourages sparsity.
# - **L2 Regularization (Ridge)**: Adds the squared magnitude of the coefficient as a penalty term to the loss function. Encourages small, non-zero coefficients.

# Q) What is a confusion matrix and how is it used?

# ## Explanation
#
# - **Confusion Matrix**: A table used to evaluate the performance of a classification model. It shows the counts of true positive, true negative, false positive, and false negative predictions.
# - **Usage**: Provides insights into the model's performance and helps calculate metrics like precision, recall, and accuracy.

# Example of Confusion Matrix
from sklearn.metrics import confusion_matrix

y_true = [0, 1, 1, 0, 1, 0, 1, 0, 1, 1]
y_pred = [0, 0, 1, 0, 1, 1, 1, 0, 0, 1]
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Q) Define AUC-ROC curve.

# ## Explanation
#
# - **AUC-ROC Curve**: The Area Under the Receiver Operating Characteristic curve. It measures the performance of a classification model at all classification thresholds.
# - **ROC Curve**: Plots the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.
# - **AUC**: Represents the degree or measure of separability. Higher AUC indicates a better performing model.

# Example of ROC Curve and AUC
from sklearn.metrics import roc_auc_score, roc_curve

y_true = [0, 0, 1, 1]
y_scores = [0.1, 0.4, 0.35, 0.8]
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
auc = roc_auc_score(y_true, y_scores)

plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Q) Explain the k-nearest neighbors algorithm.

# ## Explanation
#
# - **k-Nearest Neighbors (k-NN)**: A non-parametric, lazy learning algorithm used for classification and regression.
# - **Basic Idea**: Classify a data point based on how its neighbors are classified.
# - **Steps**:
#   1. Choose the number of neighbors, k.
#   2. Calculate the distance between the new data point and all training data points.
#   3. Select the k-nearest neighbors to the new data point.
#   4. Assign the class by majority vote (classification) or average (regression) of the k-nearest neighbors.

# Q) Explain the basic concept of a Support Vector Machine (SVM).

# ## Explanation
#
# - **Support Vector Machine (SVM)**: A supervised learning model used for classification and regression.
# - **Basic Concept**: Finds the hyperplane that best separates the data into classes.
# - **Key Components**:
#   - **Support Vectors**: Data points that are closest to the hyperplane and influence its position.
#   - **Margin**: The distance between the hyperplane and the nearest support vector. SVM aims to maximize this margin.

# Q) How does the kernel trick work in SVM?

# ## Explanation
#
# - **Kernel Trick**: A technique used to transform data into a higher dimension to make it linearly separable.
# - **How It Works**: Uses a kernel function to compute the dot product in the transformed space without explicitly transforming the data.
# - **Common Kernels**:
#   - Linear
#   - Polynomial
#   - Radial Basis Function (RBF)
#   - Sigmoid

# Q) What are the different types of kernels used in SVM and when would you use each?

# ## Explanation
#
# - **Linear Kernel**: Use when data is linearly separable.
# - **Polynomial Kernel**: Use when data is not linearly separable and the relationship is polynomial.
# - **Radial Basis Function (RBF) Kernel**: Use when data is not linearly separable and the relationship is complex. This is the most common kernel.
# - **Sigmoid Kernel**: Use in neural networks.

# Q) What is the hyperplane in SVM and how is it determined?

# ## Explanation
#
# - **Hyperplane**: A decision boundary that separates different classes in the feature space.
# - **Determination**: SVM finds the hyperplane that maximizes the margin between the classes. This is done using optimization techniques to solve for the weights and bias that define the hyperplane.

# Q) What are the pros and cons of using a Support Vector Machine (SVM)?

# ## Pros:
# - Effective in high-dimensional spaces.
# - Memory efficient as it uses support vectors.
# - Versatile with different kernel functions.
#
# ## Cons:
# - Not suitable for large datasets due to high computational cost.
# - Less effective when classes overlap.
# - Choosing the right kernel can be difficult.

# Q) Explain the difference between a hard margin and a soft margin SVM.

# ## Explanation
#
# - **Hard Margin SVM**: No misclassification is allowed. Suitable when data is linearly separable without any errors.
# - **Soft Margin SVM**: Allows some misclassification to achieve better generalization. Introduces a penalty for misclassification to handle non-linearly separable data.

# Q) Describe the process of constructing a decision tree.

# ## Explanation
#
# - **Process**:
#   1. **Select Attribute**: Choose the best attribute to split the data based on a criterion (e.g., information gain).
#   2. **Create Node**: Create a decision node based on the selected attribute.
#   3. **Split Data**: Split the dataset into subsets based on the attribute's values.
#   4. **Repeat**: Repeat the process for each subset until reaching a stopping criterion (e.g., maximum depth or pure nodes).

# Q) Describe the working principle of a decision tree.

# ## Explanation
#
# - **Working Principle**:
#   - Starts from the root node and makes decisions based on attribute values.
#   - Splits data recursively at each node based on the best attribute.
#   - Continues until reaching leaf nodes, which represent class labels or continuous values.

# Q) What is information gain and how is it used in decision trees?

# ## Explanation
#
# - **Information Gain**: A measure of the reduction in entropy or impurity after a dataset is split based on an attribute.
# - **Usage**: Used to select the attribute that best separates the data at each node. The attribute with the highest information gain is chosen for splitting.

# Q) Explain Gini impurity and its role in decision trees.

# ## Explanation
#
# - **Gini Impurity**: A measure of how often a randomly chosen element would be incorrectly classified. It ranges from 0 (pure) to 0.5 (impure).
# - **Role**: Used as a criterion to choose the attribute for splitting the data. The attribute that reduces the Gini impurity the most is selected.

# Q) What are the advantages and disadvantages of decision trees?

# ## Advantages:
# - Easy to understand and interpret.
# - Can handle both numerical and categorical data.
# - Requires little data preprocessing.
#
# ## Disadvantages:
# - Prone to overfitting.
# - Can be unstable with small variations in data.
# - Greedy algorithms may not produce the optimal tree.

# Q) How do random forests improve upon decision trees?

# ## Explanation
#
# - **Random Forests**: An ensemble learning method that combines multiple decision trees to improve performance.
# - **Improvements**:
#   - Reduces overfitting by averaging multiple trees.
#   - Increases accuracy and robustness.
#   - Reduces variance compared to individual decision trees.

# Q) How does a random forest algorithm work?

# ## Explanation
#
# - **Process**:
#   1. **Bootstrap Sampling**: Create multiple subsets of the training data using sampling with replacement.
#   2. **Train Trees**: Train a decision tree on each subset.
#   3. **Aggregation**: Combine the predictions of all trees by majority vote (classification) or averaging (regression).
```

```python
# Q) What is bootstrapping in the context of random forests?

# ## Explanation
#
# - **Bootstrapping**: A resampling technique where multiple subsets of the original dataset are created by sampling with replacement.
# - **In Random Forests**: Each decision tree is trained on a different bootstrapped subset of the data. This introduces diversity among the trees and helps reduce overfitting.

# Q) Explain the concept of feature importance in random forests.

# ## Explanation
#
# - **Feature Importance**: A measure of how much each feature contributes to the model's predictions.
# - **In Random Forests**: Calculated by measuring the decrease in impurity (e.g., Gini impurity) or the average increase in model accuracy when a feature is used for splitting.

# Q) What are the key hyperparameters of a random forest and how do they affect the model?

# ## Key Hyperparameters:
# - **n_estimators**: Number of trees in the forest. More trees generally improve performance but increase computation time.
# - **max_depth**: Maximum depth of each tree. Controls overfitting; deeper trees can capture more complexity but may overfit.
# - **min_samples_split**: Minimum number of samples required to split an internal node. Higher values prevent overfitting by requiring more samples to create a split.
# - **min_samples_leaf**: Minimum number of samples required to be at a leaf node. Higher values prevent overfitting by making leaves more generalized.
# - **max_features**: Number of features to consider when looking for the best split. Smaller values reduce variance but increase bias.

# Q) Describe the logistic regression model and its assumptions.

# ## Explanation
#
# - **Logistic Regression**: A statistical model used for binary classification. It models the probability of a binary outcome using the logistic function.
# - **Assumptions**:
#   - The outcome is binary.
#   - Observations are independent.
#   - There is no multicollinearity among the predictors.
#   - The relationship between the independent variables and the log odds of the dependent variable is linear.

# Q) How does logistic regression handle binary classification problems?

# ## Explanation
#
# - **Binary Classification**: Logistic regression models the probability that a given input belongs to a particular class.
# - **Process**:
#   1. Computes the linear combination of input features and weights.
#   2. Applies the sigmoid function to map the output to a probability.
#   3. Classifies the input as the class with the highest probability.

# Q) What is the sigmoid function and how is it used in logistic regression?

# ## Explanation
#
# - **Sigmoid Function**: A mathematical function that maps any real-valued number into the range (0, 1).
# - **Formula**: `σ(x) = 1 / (1 + e^(-x))`
# - **Usage in Logistic Regression**: Transforms the linear combination of inputs and weights into a probability value between 0 and 1.

# Q) Explain the concept of the cost function in logistic regression.

# ## Explanation
#
# - **Cost Function**: Measures the error between the predicted probabilities and the actual class labels.
# - **Logistic Regression Cost Function**: Also known as the log-loss or binary cross-entropy.
# - **Formula**: `J(θ) = -1/m * Σ [y*log(hθ(x)) + (1-y)*log(1-hθ(x))]`
# - **Purpose**: Used to find the optimal parameters (weights) that minimize the prediction error.

# Q) How can logistic regression be extended to handle multiclass classification?

# ## Explanation
#
# - **One-vs-Rest (OvR)**: Fits a separate binary classifier for each class, where the class is treated as one class against all other classes.
# - **One-vs-One (OvO)**: Fits a separate binary classifier for each pair of classes and uses voting to determine the final class.
# - **Softmax Regression (Multinomial Logistic Regression)**: Generalizes logistic regression to multiple classes by using the softmax function to compute probabilities.

# Q) What is the difference between L1 and L2 regularization in logistic regression?

# ## Explanation
#
# - **L1 Regularization (Lasso)**:
#   - Adds the absolute value of the coefficients as a penalty term to the cost function.
#   - Promotes sparsity, leading to feature selection.
#   - Formula: `J(θ) = J(θ) + λ * Σ|θ_j|`
#
# - **L2 Regularization (Ridge)**:
#   - Adds the squared value of the coefficients as a penalty term to the cost function.
#   - Prevents overfitting by shrinking coefficients.
#   - Formula: `J(θ) = J(θ) + λ * Σθ_j^2`

# Q) What is XGBoost and how does it differ from other boosting algorithms?

# ## Explanation
#
# - **XGBoost**: An optimized implementation of the gradient boosting algorithm designed for speed and performance.
# - **Differences**:
#   - Uses regularization to prevent overfitting.
#   - Employs parallel processing for faster computation.
#   - Incorporates handling of missing values.
#   - Provides built-in cross-validation.

# Q) Explain the concept of boosting in the context of ensemble learning.

# ## Explanation
#
# - **Boosting**: An ensemble learning technique that combines multiple weak learners (e.g., decision trees) to create a strong learner.
# - **Process**:
#   1. Train a weak learner on the data.
#   2. Adjust the weights of misclassified examples to focus on harder cases.
#   3. Train the next weak learner on the adjusted data.
#   4. Repeat the process and combine the predictions of all learners.

# Q) How does XGBoost handle missing values?

# ## Explanation
#
# - **Handling Missing Values**:
#   - XGBoost automatically learns the best direction (left or right) to handle missing values during tree construction.
#   - It treats missing values as a separate category and finds the optimal split to minimize loss.

# Q) What are the key hyperparameters in XGBoost and how do they affect model performance?

# ## Key Hyperparameters and their Effects:
#
# - **n_estimators**: Number of trees in the ensemble. More trees generally improve performance but increase computation time.
# - **learning_rate**: Shrinks the contribution of each tree. Lower values require more trees but can improve performance.
# - **max_depth**: Maximum depth of each tree. Controls model complexity; deeper trees can capture more details but may overfit.
# - **min_child_weight**: Minimum sum of instance weights needed in a child. Higher values prevent overfitting by making trees more generalized.
# - **gamma**: Minimum loss reduction required to make a further partition. Higher values prevent overfitting by creating more conservative splits.
# - **subsample**: Fraction of samples used for fitting each tree. Lower values prevent overfitting but can increase bias.
# - **colsample_bytree**: Fraction of features used for fitting each tree. Lower values prevent overfitting but can increase bias.
# - **lambda**: L2 regularization term on weights. Prevents overfitting by shrinking weights.
# - **alpha**: L1 regularization term on weights. Promotes sparsity in the model.

# Q) Describe the process of gradient boosting in XGBoost.

# ## Explanation
#
# - **Gradient Boosting Process**:
#   1. **Initialize** the model with a constant value (e.g., the mean of the target variable).
#   2. **Iteratively**:
#      - Compute the **residuals** (errors) between the current model predictions and the actual values.
#      - Fit a **weak learner** (e.g., a decision tree) to the residuals.
#      - **Update** the model by adding the predictions of the weak learner, scaled by a learning rate.
#   3. **Combine** the predictions of all learners to make the final prediction.

# Q) What are the advantages and disadvantages of using XGBoost?

# ## Advantages:
# - **High Performance**: XGBoost is optimized for speed and accuracy.
# - **Regularization**: Built-in L1 and L2 regularization help prevent overfitting.
# - **Handling Missing Values**: Automatically handles missing values during training.
# - **Parallel Processing**: Supports parallel computation, making it faster than other implementations.
# - **Flexibility**: Can handle various objective functions and evaluation metrics.
# - **Feature Importance**: Provides insights into feature importance.
#
# ## Disadvantages:
# - **Complexity**: Can be more complex to tune compared to simpler models.
# - **Computationally Intensive**: Requires significant computational resources for large datasets.
# - **Overfitting**: Despite regularization, can still overfit on noisy data if not tuned properly.
```
